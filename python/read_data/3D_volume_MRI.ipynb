```python
import SimpleITK as sitk
from matplotlib import pyplot as plt
```

```python
#file path
fn_mri = "../data/T2_MRI.nii.gz"
fn_slice = "../data/T2_MRI_10.png"
```

```python
#read the files and show the size and pixel dimensions
im_mri = sitk.ReadImage(fn_mri)
print(im_mri.GetSize(), im_mri.GetSpacing())
```

```python
slice_idx = 10
slice_mri = im_mri[:,:,slice_idx]
slice_mri_arr = sitk.GetArrayFromImage(slice_mri)
plt.imshow(slice_mri_arr, cmap='gray'); plt.show()
```

```python
#write only one slice
im_mri_rescaled = sitk.Cast(sitk.RescaleIntensity(im_mri,0,255),sitk.sitkUInt8)
sitk.WriteImage(im_mri_rescaled[:,:,slice_idx], fn_slice)
```
